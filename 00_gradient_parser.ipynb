```python
import numpy as np
import pickle
file = "cifar_cnn_users[10]_rounds[100]_frac[1.0]_iid[1]_local_ep[10]_local_bs[10]_attck_frac[0.2]_grads.pkl"
with open(file,"rb") as f:
    data = pickle.load(f)
```

```python
#Grads data structure
print("\nGrads datastructure\n")
for r in range(len(data)):
    #Iterate through users/participants in current round
    print("round: ",r)
    for idx in data[r]:
        print("\tuser: ",idx,"\tgrads:", data[r][idx][0].keys(),"\tlabel:",data[r][idx][1])
```

```python
#Calculate cos_sim matrix for conv1

def get_round_euclidean_distance_matrix(data,layer):
    matrix = [[0 for i in range(len(data))] for i in range(len(data))]
    for user1,(X1,y1) in data.items():
        for user2,(X2,y2) in data.items():
            dist = np.linalg.norm((X1[layer] - (X2[layer])))
            matrix[user1][user2] = round(dist,4)
    return matrix

def get_round_cos_sim_matrix(data,layer):
    matrix = [[0 for i in range(len(data))] for i in range(len(data))]
    for user1,(X1,y1) in data.items():
        for user2,(X2,y2) in data.items():
            x1 = np.array(X1[layer]).flatten()
            x2 = np.array(X2[layer]).flatten()
            norm_product = np.linalg.norm(x1) * np.linalg.norm(x2)
            cos_sim = np.dot(x1,x2)/ norm_product if norm_product != 0 else 0
            matrix[user1][user2] = round(cos_sim,4)
    return matrix

matrix1 = get_round_euclidean_distance_matrix(data[0],'conv1')
matrix2 = get_round_cos_sim_matrix(data[0],'conv1')

print("Conv1 euclidean distance matrix\n")
for row in matrix1:
    print(row)
print("\nConv1 cosine similarity matrix\n")
for row in matrix2:
    print(row)
```

```python
#Check cos_sim for conv1 layer for all rounds, create cos_sim dataset and labels
print("\nConv1 cosine similarity matrix for all rounds\n")
X, y = [],[]
for r in range(len(data)):
    print("round: ",r)
    matrix = get_round_cos_sim_matrix(data[0],'conv1')
    for i,row in enumerate(matrix):
        #del row[i]
        X.append(row)
        y.append(1 if i == 5 or i == 6 else 0)
        print("\t",X[-1])
```

```python
#Check cos_sim for conv1 layer for all rounds
print("\nConv1 euclidean distance matrix for all rounds\n")
for r in range(len(data)):
    print("round: ",r)
    matrix = get_round_euclidean_distance_matrix(data[0],'conv1')
    for row in matrix:
        print("\t",row)
```

```python
#Format data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1234)

X_train = torch.from_numpy(np.array(X_train).astype(np.float32))
y_train = torch.from_numpy(np.array(y_train).astype(np.float32))
X_test = torch.from_numpy(np.array(X_test).astype(np.float32))
y_test = torch.from_numpy(np.array(y_test).astype(np.float32))

print(len([1 for d in y_train if d == 1]))

y_train = y_train.view(y_train.shape[0],1)
y_test = y_test.view(y_test.shape[0],1)
```

```python
#GradNet model for classifying gradients
import torch
import torch.nn as nn
import torch.nn.functional as F

# Hyper-parameters 
num_epochs = 1000
batch_size = 10
learning_rate = 0.0001

class GradNet(nn.Module):
    
    def __init__(self,input_size):
        super(GradNet,self).__init__()
        self.linear1 = nn.Linear(input_size,1)
    
    def forward(self,X):
        return torch.sigmoid(self.linear1(X))

model = GradNet(10)
weight = torch.tensor([0.23])
criterion = nn.BCELoss(weight=weight)
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

```python
#Training
for epoch in range(num_epochs):
    # Forward pass
    y_pred = model(X_train)

    loss = criterion(y_pred, y_train)
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
    if (epoch + 1) % 100 == 0:
        print(f'epoch: {epoch+1} \t loss = {loss.item()}')
        
```

```python
#Evaluation
from sklearn.metrics import confusion_matrix

def performance(X_test,y_test):
    with torch.no_grad(): #Similar to detach, defuses gradient storage for the model
        y_pred = model(X_test)
        y_pred_classes = y_pred.round() #round preds to 0 or 1
        acc = y_pred_classes.eq(y_test).sum() / float(y_test.shape[0])
        print(confusion_matrix(y_test, y_pred_classes))
        tn, fp, fn, tp = confusion_matrix(y_test, y_pred_classes).ravel()
        print("TP: ",tp)
        print("TN: ",tn)
        print("FP: ",fp)
        print("FN: ",fn)
        print(f'Acc: {acc:.4f}')

print("+++ Train +++\n")
performance(X_train,y_train)
print("\n+++ Test +++\n")
performance(X_test,y_test)
```
